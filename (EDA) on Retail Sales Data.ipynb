```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Data Loading and Cleaning

```python
df = pd.read_csv('retail_sales_dataset.csv')
```

```python
df.head()
```

#### Descriptive Statistics

```python
mean_value = df['Total Amount'].mean()
print("Mean:", mean_value)
```

```python
mean_value = df['Total Amount'].median()
print("Median:", mean_value)
```

```python
mean_value = df['Total Amount'].mode()[0]
print("Mode:", mean_value)
```

```python
mean_value = df['Total Amount'].std()
print("std:", mean_value)
```

#### Time Series Analysis

```python
plt.figure(figsize=(6,4))
plt.plot(df['Total Amount'], label='Total Amount')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.legend()
plt.show()
```

#### Customer and Product Analysis

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### Visualization

```python
product_count = df['Product Category'].value_counts()

# Plot the top most frequently purchased products
plt.figure(figsize=(8,4))
product_count.head(10).plot(kind='bar', color='orange')
plt.title('Top Most Purchased Products')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
print(df)
```

```python
sns.lineplot(x="Quantity", y="year", data=df)
plt.title("Total Quantity by Month")
plt.xlabel("Month of the Week")
plt.ylabel("Total Qunatity")
plt.show()
```

```python
corr = df[['Age', 'Total Amount', 'Quantity']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Customer Demographics and Spending')
plt.show()
```

### Recommendation

####  Target Specific Customer Demographics

##### Age Segmentation: If a large portion of your customers fall into a specific age group, target this group with products that align with their interests. For example, if the majority of your customers are in the 25-50 age range, consider offering products that appeal to young professionals, such as tech gadgets or lifestyle accessories

##### Income Segmentation: For high-income customers, promote premium or luxury products, while for lower-income segments, focus on budget-friendly or discount-oriented offerings

##### Gender Targeting: If your analysis shows that a certain gender (e.g., women) tends to purchase more frequently, consider creating gender-specific campaigns or product bundles.

#### Increase Average Order Value (AOV)


#####  Upsell and Cross-Sell: Promote complementary or higher-value products during the checkout process (e.g., "Customers who bought this also bought..."). This can increase AOV by encouraging customers to purchase more.

##### Free Shipping Threshold: Introduce a free shipping threshold (e.g., free shipping for orders over $50). This encourages customers to add more items to their cart to qualify for the offer.

##### Discount on Minimum Spend: Offer discounts for customers who spend above a certain amount. For example, "Spend $100 and get 10% off" to increase their total order size.
